 # Exploratory Data Analysis for Diabetes Health Indicators

### Data source :https://archive.ics.uci.edu/dataset/891/cdc+diabetes+health+indicators

```python
#import statements
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_Path="/Users/sathvikchava/my files/big data/cdc_diabetes_health_indicators.csv"
df=pd.read_csv(file_Path)
```

```python
df.head()
```

### Data Overview 

```python
print(" Data Set Information ")
print(df.info())
```

```python
# finding missing values
df.isnull().sum()
```

##### No missing values found 

```python
# Summary Statistics
df.describe()
```

#### most of the columns are catogorical and we are being very careful while drawing any conclusions from stats


```python
# Analysis Target variable 
plt.figure(figsize=(6,4))
df['Diabetes_binary'].value_counts(normalize=True).plot(kind='bar')
plt.title("Target Variable Distribuiton")
plt.xlabel('Diabetes_binary')
plt.ylabel('Proportion')
plt.show()
                                                        
```

```python
# Numerical feature analysis 

df['BMI'].hist(bins=30, figsize=(12, 8))
plt.show()
```

```python
# Other categorical features which has more than 2 catogories
df['MentHlth'].hist(bins=30, figsize=(12, 8))
plt.show()
```

```python
df['PhysHlth'].hist(bins=30, figsize=(12, 8))
plt.show()
```

```python
df['Age'].hist(bins=30, figsize=(12, 8))
plt.show()
```

```python
# Heat Map for all 


correlation_matrix = df.corr()

# Correlation of varibales with target variable
correlation_values = correlation_matrix['Diabetes_binary'].sort_values(ascending=False)
print("\n Correlations in sorted order")
print(correlation_values)
```

```python



plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', cbar=True)
plt.title(" Heatmap for correlations between every varibale")
plt.show()
```
