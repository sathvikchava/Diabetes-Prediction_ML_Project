```python
# import statements 
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# connecting to dataset
file_Path="/Users/sathvikchava/my files/big data/cdc_diabetes_health_indicators.csv"
df=pd.read_csv(file_Path)
```

```python
df.head()
```

```python
# Seperatng binary,non binary and numerical features features 

# Selecting columns with only 2 unique values
binary_Features= df.columns[(df.nunique() == 2)]
binary_Features
```

```python
numerical_Features= ['BMI']
numerical_Features = pd.Index(numerical_Features)
numerical_Features
```

```python
non_binary_Features=df.columns.difference(binary_Features.union(numerical_Features))
non_binary_Features
```

```python
# Doing One-Hot encoding 
df=pd.get_dummies(df,columns=non_binary_Features,drop_first=True)
```

```python
# feature scaling 
scaler=StandardScaler()
df[numerical_Features]=scaler.fit_transform(df[numerical_Features])
```

```python
df.head()
```

```python
df.iloc[:, 6:] = df.iloc[:, 6:].replace({True: 1, False: 0})
```

```python
df.to_csv('/Users/sathvikchava/my files/big data/Project files/Diabetes_Health_Indicators_cleanedData1.0.csv', index=False)
```
